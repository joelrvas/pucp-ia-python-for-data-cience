```python
import pandas as pd

# Cargar los datos desde los archivos CSV
df_ubigeos = pd.read_csv('TB_UBIGEOS.csv', delimiter=';')
df_vacunacion = pd.read_csv('TB_CENTRO_VACUNACION.csv', delimiter=';')

# Verificar si hay datos faltantes en 'entidad_administra' en df_vacunacion
print(df_vacunacion[['id_ubigeo', 'entidad_administra']].isna().sum())

# Unir los datos de vacunación y ubigeos
df_join_ubigeo_centros = pd.merge(df_vacunacion, df_ubigeos, how='left', left_on='id_ubigeo', right_on='id_ubigeo')


# Verificar si hay duplicidad de columnas
print(df_join_ubigeo_centros.columns)

# Asegurarse de que el id_ubigeo sea del mismo tipo en ambos DataFrames
df_join_ubigeo_centros['id_ubigeo'] = df_join_ubigeo_centros['id_ubigeo'].astype(str)

# Filtrar por id_ubigeo
filtro = '31'
resultado_centro_vacunacion = df_join_ubigeo_centros.loc[df_join_ubigeo_centros['id_ubigeo'] == filtro]

# Mostrar los resultados filtrados
resultado_centro_vacunacion.head()
```
